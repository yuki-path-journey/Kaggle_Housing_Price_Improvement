```python
import pandas as pd
train = pd.read_csv('train.csv')
missing_values = train.isnull().sum().sort_values(ascending=False)
print(missing_values.head(20)) # 上位20項目を表示
```

```python
train['PoolQC'] = train['PoolQC'].fillna('None')
print(train['PoolQC'].isnull().sum())
```

```python
train['MiscFeature'] = train['MiscFeature'].fillna('None')
print(f"MiscFeatureの残り欠損数: {train['MiscFeature'].isnull().sum()}")
```

```python
train['Alley'] = train['Alley'].fillna('None')
print(f"Alleyの残り欠損数: {train['Alley'].isnull().sum()}")
```

```python
train['Fence'] = train['Fence'].fillna('None')
print(f"Fenceの残り欠損数: {train['Fence'].isnull().sum()}")
```

```python
train['MasVnrType'] = train['MasVnrType'].fillna('None')
print(f"MasVnrTypeの残り欠損数: {train['MasVnrType'].isnull().sum()}")
```

```python
train['FireplaceQu'] = train['FireplaceQu'].fillna('None')
print(f"FireplaceQuの残り欠損数: {train['FireplaceQu'].isnull().sum()}")
```

```python
train['LotFrontage'] = train['LotFrontage'].fillna(train['LotFrontage'].median())
print(f"LotFrontagenの残り欠損数: {train['LotFrontage'].isnull().sum()}")
```

```python
garage_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cols:
    train[col] = train[col].fillna('None')
train['GarageYrBlt'] = train['GarageYrBlt'].fillna(0)
print(train.isnull().sum().sort_values(ascending=False).head(10))
```

```python
bsmt_cols = ['BsmtExposure', 'BsmtFinType2', 'BsmtCond', 'BsmtQual', 'BsmtFinType1']
for col in bsmt_cols:
    train[col] = train[col].fillna('None')
train['MasVnrArea'] = train['MasVnrArea'].fillna(0)
train['Electrical'] = train['Electrical'].fillna(train['Electrical'].mode()[0])
print("全データの欠損値合計:", train.isnull().sum().sum())
```

```python
train.to_csv('train_cleaned.csv', index=False)
```

```python
import pandas as pd
train = pd.read_csv('train.csv')
none_cols = [
    'PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType',
    'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
    'BsmtExposure', 'BsmtFinType2', 'BsmtCond', 'BsmtQual', 'BsmtFinType1'
]
for col in none_cols:
    train[col] = train[col].fillna('None')
train['LotFrontage'] = train['LotFrontage'].fillna(train['LotFrontage'].median())
train['MasVnrArea'] = train['MasVnrArea'].fillna(0)
train['GarageYrBlt'] = train['GarageYrBlt'].fillna(0)
train['Electrical'] = train['Electrical'].fillna(train['Electrical'].mode()[0])
print(f"残りの欠損値合計: {train.isnull().sum().sum()}")
train.to_csv('train_cleaned.csv', index=False)
```
